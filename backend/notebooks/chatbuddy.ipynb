```python
from langgraph.graph import StateGraph, START, END
from langchain_core.messages import BaseMessage, HumanMessage, AIMessage
from typing import TypedDict, Annotated
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.graph.message import add_messages
from pydantic import BaseModel
from dotenv import load_dotenv

load_dotenv()
```

```python
class ChatBuddyState(TypedDict):
    messages: Annotated[list[BaseMessage], add_messages, "A list of messages in the conversation"]
```

```python
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_methods =['*'],
    allow_headers=['*'],
    allow_credentials=True
)

class ChatRequest(BaseModel):
    user_message: str


@app.get("/api/")
def greet():
    return {"success": True, "message": "Hello from the API!"} 


@app.post("/api/chat")
def chatting_buddy(request: ChatRequest):
    try:
        query = request.user_message
        initial_state = {}
        response = workflow.invoke()
        
    intial_state = {
        "message": [HumanMessage(content = query)]
    }
```

```python
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.9)
```

```python
def chat_node(state:ChatBuddyState) -> ChatBuddyState:
    message = state["message"]
    response = model.invoke(message)
    return {"message": response}
```

```python
# Create workflow
graph = StateGraph(ChatBuddyState)

graph.add_node("chat_node", chat_node)

graph.add_edge(START, "chat_node")
graph.add_edge("chat_node", END)

workflow = graph.compile()
```
